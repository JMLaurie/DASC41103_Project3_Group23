```python
"""
Create a dataset from a list of values from 0 to 50
"""
# create dataset
t = torch.arange(51, dtype=torch.float32)
print(f"Torch t: {t}")
print("-"*50)

# required library
from torch.utils.data import DataLoader

# initialize the DataLoader class with data from torch object, t
data_loader = DataLoader(t)

# print items in data_loader
for item in data_loader:
    print(item)
print("-"*50)

# create batches from dataset with desired batch_size (e.g., number of elements)
# batch_szie = 1 produces the data_loader above
# creates 3 batches
data_loader = DataLoader(t, batch_size=3)

# print items in data_loader
for item in data_loader:
    print(item)
print("-"*50)
```

```python
"""
Creating a joint dataset by combining two tensors
Using our created Dataset class
"""
# create data
torch.manual_seed(1)
t_x = torch.rand([51, 3], dtype=torch.float32)
t_y = torch.arange(51)
print(f"Torch t_x: {t_x}")
print(f"Torch t_y: {t_y}")
print("-"*50)

# required library to use Dataset
from torch.utils.data import Dataset

# Create a user defined function Dataset class
# by using the "duner" or double-underscore to create methods we are creating special methods in Python that
#     python recognizes to call them automaticaly in response to built-in operations, rather than invoking them explicityly ourselves
class Joint_UDF_Dataset(Dataset):
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def __len__(self):
        return len(self.x)
    # returns the corresponding sample to the given index
    def __getitem__(self, index):
        return self.x[index], self.y[index]

# create joint dataset of t_x and t_y with user defined class of Dataset
joint_dataset = Joint_UDF_Dataset(t_x, t_y)

# print items in joint_dataset
for item in joint_dataset:
    print(' x: ', item[0], ' y: ', item[1])
```

```python
"""
Creating a joint dataset by using torch.utils.data Dataset class called TensorDataset
"""
from torch.utils.data import TensorDataset
# create joint dataset of t_x and t_y with TensorDataset class
# error in book..states to use JointDataset class but code block uses TensorDataset
joint_tensor_dataset = TensorDataset(t_x, t_y)

# print items in joint_dataset
for item in joint_tensor_dataset:
    print(' x: ', item[0], ' y: ', item[1])
```

```python
"""
Shuffle, batch, and repeat with dataLoaders
Using TensorDataset merged data from above
"""
# required library
from torch.utils.data import DataLoader
torch.manual_seed(1)

# loads joint_tensor_dataset into data_loader as batches and with shuffling
# shuffles rows without losing 1 to 1 correspondence between the entries in x and y.
# can compare above data with data printed below
# with shuffle=True, each time you iterate through the DataLoader for a new training epoch, you will receive a completely different set of randomly ordered batches
data_loader = DataLoader(joint_tensor_dataset, batch_size=8, shuffle=True)

# print items in data_loader
for item in data_loader:
    print(item)
print("-"*50)

# print items in joint_tensor_dataset by batch
# the 1 in enumerate is the optional start parameter to start index at 1 instead of the default of 0
for i, batch in enumerate(data_loader, 1):
    print(f'batch {i}:', 'x:', batch[0], '\n y:', batch[1])
print("-"*50)

# shuffle and iterate over the dataset by a desired number of epochs
for epoch in range(2):
    # epoch + 1 is required since starts at 0
    print(f'epoch {epoch + 1}:')
    for i, batch in enumerate(data_loader, 1):
        print(f'batch {i}:', 'x:', batch[0], '\n y:', batch[1])
```

```python
# transform data from path with response variable to pixel data with response variable
# first we need to create a transformation
# required libraries
import torchvision.transforms as transforms
# set a standard aspect ratio for all images (# of pixels by # of pixels) to resize to
img_height, img_width = 500, 500
# defines a transform pipeline using torchvision.transforms.Compose
# compose allows chaining multiple transformations together
# when an image is passed to the transform object, the transformation in Compose is executed in the order provided in the list
#      .ToTensor() converts a PIL image or Numpy ndarray to a Tensor
#           image becomes a torch.FloatTensor that scales original pixel value [0, 255] to [0, 1] scale and rearranges dimensions from (H, W, C) to PyTorch standard (C, H, W)
#      .Resize() resizes image to the specified dimensions to ensure a consistent size which is often requirement for deep learning models (makes it easier)
transform = transforms.Compose([
    transforms.Lambda(lambda img: img.convert("RGB")),      # <- this ensures 3 channels
    transforms.Resize((img_height, img_width)),
    transforms.ToTensor(),
])
# update our user defined function class for joining to include transformation
# gets the transformed pixel information and join with response label instead of combining file name/path with response variable
class ImageDataset(Dataset):
    def __init__(self, file_list, labels, transform=None):
        self.file_list = file_list
        self.labels = labels
        self.transform = transform
    def __getitem__(self, index):
        img = Image.open(self.file_list[index])
        # Ensure all images are RGB
        img = img.convert('RGB')
        if self.transform is not None:
            img = self.transform(img)
        label = self.labels[index]
        return img, label
    def __len__(self):
        return len(self.labels)
# initialize perform join with data
image_dataset = ImageDataset(file_list, labels, transform)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Subset
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# Check for GPU availability. 
# If a GPU is available (e.g., on Databricks or a local machine with NVIDIA), 
# we set the device to 'cuda'. Otherwise, we default to 'cpu'.
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# Define image dimensions for consistency across the model
img_height, img_width = 500, 500  
```

```python
# --- Data Preparation ---

# Create a list of indices (0 to total_images - 1)
indices = np.arange(len(image_dataset))

# Split indices: 80% for training, 20% for validation.
# random_state=1 ensures we get the same split every time we run the code.
train_idx, val_idx = train_test_split(indices, test_size=0.2, random_state=1)

# Create Subset objects based on those indices
train_ds = Subset(image_dataset, train_idx)
val_ds = Subset(image_dataset, val_idx)

# Define a default batch size (will be overridden in tuning later, but good for testing)
batch_size = 32  

# DataLoaders handle batching and shuffling.
# shuffle=True for training helps the model generalize better by seeing data in different orders.
# shuffle=False for validation ensures consistent evaluation metrics.
train_dl = DataLoader(train_ds, batch_size=batch_size, shuffle=True)
val_dl = DataLoader(val_ds, batch_size=batch_size, shuffle=False)
```

```python
class RazorbackCNN(nn.Module):
    def __init__(self, in_channels=3, img_height=500, img_width=500):
        super().__init__()
        
        # --- Feature Extractor ---
        # A sequence of Convolution -> ReLU (activation) -> MaxPool (downsampling)
        self.features = nn.Sequential(
            # Layer 1: Input 3 channels (RGB) -> Output 32 feature maps
            nn.Conv2d(in_channels, 32, kernel_size=3, padding=1),
            nn.ReLU(),       # Adds non-linearity
            nn.MaxPool2d(2), # Reduces image size by half (250x250)
            nn.Dropout(0.5), # Prevents overfitting by randomly zeroing out neurons

            # Layer 2: 32 -> 64 feature maps
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2), # Reduces size by half again (125x125)
            nn.Dropout(0.5),

            # Layer 3: 64 -> 128 feature maps
            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2), # Reduces size (~62x62)

            # Layer 4: 128 -> 256 feature maps
            nn.Conv2d(128, 256, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2)  # Reduces size (~31x31)
        )
        
        # --- Automatic Flattening Calculation ---
        # Instead of manually calculating the input size for the Linear layer,
        # we pass a "dummy" tensor through the feature layers to see the output shape.
        with torch.no_grad():
            dummy_input = torch.zeros(1, in_channels, img_height, img_width)
            # Get the number of flattened features coming out of the CNN block
            n_feats = self.features(dummy_input).view(1, -1).shape[1]
        
        # --- Binary Classifier ---
        self.classifier = nn.Sequential(
            nn.Flatten(),           # Turn the 3D tensor into a 1D vector
            nn.Linear(n_feats, 1),  # Reduce to a single output node
            nn.Sigmoid()            # Squash output between 0 and 1 (probability)
        )

    def forward(self, x):
        # Pass input through features, then classifier
        x = self.features(x)
        x = self.classifier(x)
        return x
```

```python
# Define the loss function for binary classification
loss_fn = nn.BCELoss()

def train_model(model, train_dl, val_dl, optimizer, epochs=10):
    """
    Training loop that runs for a specific number of epochs.
    Returns a dictionary 'hist' containing loss and accuracy for plotting.
    """
    hist = {'train_loss':[], 'val_loss':[], 'train_acc':[], 'val_acc':[]}
    
    for epoch in range(epochs):
        # --- Training Phase ---
        model.train() # Set model to training mode (enables Dropout)
        train_loss, train_corrects, n_train = 0, 0, 0
        
        for xb, yb in train_dl:
            # Move data to GPU/CPU and ensure labels are correct shape
            xb, yb = xb.to(device), yb.float().unsqueeze(1).to(device)
            
            pred = model(xb)             # Forward pass
            loss = loss_fn(pred, yb)     # Calculate loss
            
            optimizer.zero_grad()        # Clear previous gradients
            loss.backward()              # Backpropagation (calc gradients)
            optimizer.step()             # Update weights
            
            # Track stats
            train_loss += loss.item() * yb.size(0)
            # Check if prediction > 0.5 (class 1) matches label
            train_corrects += ((pred > 0.5).float() == yb).sum().item()
            n_train += yb.size(0)
            
        # --- Validation Phase ---
        model.eval() # Set to eval mode (disables Dropout)
        val_loss, val_corrects, n_val = 0, 0, 0
        
        with torch.no_grad(): # No need to track gradients for validation
            for xb, yb in val_dl:
                xb, yb = xb.to(device), yb.float().unsqueeze(1).to(device)
                pred = model(xb)
                loss = loss_fn(pred, yb)
                
                val_loss += loss.item() * yb.size(0)
                val_corrects += ((pred > 0.5).float() == yb).sum().item()
                n_val += yb.size(0)
        
        # --- Calculate Epoch Averages ---
        epoch_train_loss = train_loss / n_train
        epoch_val_loss = val_loss / n_val
        epoch_train_acc = train_corrects / n_train
        epoch_val_acc = val_corrects / n_val
        
        # Store history
        hist['train_loss'].append(epoch_train_loss)
        hist['val_loss'].append(epoch_val_loss)
        hist['train_acc'].append(epoch_train_acc)
        hist['val_acc'].append(epoch_val_acc)
        
        print(f"Epoch {epoch+1}/{epochs} - Train loss: {epoch_train_loss:.4f}, acc: {epoch_train_acc:.3f} | Val loss: {epoch_val_loss:.4f}, acc: {epoch_val_acc:.3f}")
        
    return hist
```

```python
# Define the grid of parameters to test
batch_sizes = [16, 32, 64]
lrs = [0.001, 0.0005]

# Variables to track the winner
best_acc = 0
best_model = None
best_hist = None
best_params = None

print("Starting Hyperparameter Tuning...")

for bs in batch_sizes:
    for lr in lrs:
        # Re-initialize DataLoaders with current batch size
        train_dl = DataLoader(train_ds, batch_size=bs, shuffle=True)
        val_dl = DataLoader(val_ds, batch_size=bs, shuffle=False)
        
        # Re-initialize the model (start fresh for every new combo)
        model = RazorbackCNN(in_channels=3, img_height=img_height, img_width=img_width).to(device)
        
        # Initialize optimizer with current learning rate
        optimizer = optim.Adam(model.parameters(), lr=lr)
        
        print(f"\nTraining with batch_size={bs}, lr={lr}:")
        
        # Train for 6 epochs (short run to test parameters)
        hist = train_model(model, train_dl, val_dl, optimizer, epochs=6)
        
        # Check if this is the best model so far
        val_acc = max(hist['val_acc'])
        if val_acc > best_acc:
            best_acc = val_acc
            best_model = model # Save the actual model object
            best_hist = hist   # Save the history for plotting
            best_params = (bs, lr)

print(f"\nWinning Config: Batch Size={best_params[0]}, LR={best_params[1]} with Acc: {best_acc*100:.2f}%")
```

```python
# --- Plotting Results ---
x_arr = np.arange(len(best_hist['train_loss'])) + 1

plt.figure(figsize=(12, 4))

# Subplot 1: Loss over time
plt.subplot(1,2,1)
plt.plot(x_arr, best_hist['train_loss'], '-o', label="Train loss")
plt.plot(x_arr, best_hist['val_loss'], '--', label="Val loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Loss Curve")
plt.legend()

# Subplot 2: Accuracy over time
plt.subplot(1,2,2)
plt.plot(x_arr, best_hist['train_acc'], '-o', label="Train acc")
plt.plot(x_arr, best_hist['val_acc'], '--', label="Val acc")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("Accuracy Curve")
plt.legend()

plt.tight_layout()
plt.show()

# --- Save Final Model ---
# We save the entire model object (architecture + weights)
torch.save(best_model, "Group_23_CNN_FullModel.pt")
print("Model successfully saved as Group_23_CNN_FullModel.pt")
```
